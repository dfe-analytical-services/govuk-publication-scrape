```python
import pandas as pd #A module to create dataframes
import requests #A module to send requests to websites
import re #A module for regular expressions
from lxml import html #A module to process website coding languages
import unicodedata #A module to format data
import timeit #A module to check how long a script takes to run
import json #A module for json files
```

```python
url ="https://www.gov.uk/government/statistics?page=1&publication_filter_option=statistics" #variable for page to be webscraped 
page = requests.get(url) #create variable page to connect to the url 
tree = html.fromstring(page.content) #define function that extracts the content from page 
noPages = tree.xpath('//*[@id="show-more-documents"]/ul/li/a/span[2]/text()')  #define function that navigates to a specific place for number of pages
noPages = int(re.sub(".*of ","", noPages[0])) #function that substracts the number of pages needed to webscrape from the path above
noPages
```

```python
publications = [] # create list for publications
counter = 0
start = timeit.default_timer()

for page_num in range(1,noPages+1):
    print str(page_num) + " start"
    url ="https://www.gov.uk/government/statistics?page=" + str(page_num) + "&publication_filter_option=statistics" #page to be webscraped 
    page = requests.get(url) #create variable page to connect to the url 
    tree = html.fromstring(page.content) #define function that extracts the content from page
    noPublications = len(tree.xpath('//ol[@class="js-document-list document-list"]/li')) #function defining number of publications per webpage

    print str(page_num) + " publications counted: " + str(noPublications)

    for j in range(1,noPublications+1):
        
        print "publication " + str(page_num) + "." + str(j)
        
        title = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/h3/a/text()')[0]  #define function that navigates to title
        link = 'https://www.gov.uk' + tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/h3/a/@href')[0] #extract link
        info_link = 'https://www.gov.uk/info' + tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/h3/a/@href')[0]
        info_page = requests.get(info_link) #create variable page to connect to the url 
        info_tree = html.fromstring(info_page.content) #define function that extracts the content from page
        
        try:
            view_count = info_tree.xpath('//*[@id="lead-metrics"]/div[1]/div[1]/p/text()')[0] #define function that navigates to count of views
            view_count = view_count.replace(" ","") #format
            view_count = view_count.replace("\n", "")
        except IndexError:
            view_count = 0

        try:
            search_count = info_tree.xpath('//*[@id="lead-metrics"]/div[2]/div/p/text()')[0] #define function that navigates to number of searches
            if "<" in search_count:
                search_count = 0
            else:
                search_count = search_count.replace(" ","") #format
                search_count = search_count.replace("\n","")
                search_count = [int(s) for s in search_count.split() if s.isdigit()][0] #extract integer
        except IndexError:
            search_count = 0
        
        try:
            problems_count = info_tree.xpath('//*[@id="lead-metrics"]/div[2]/div/p/text()')[0] #define function that navigates to number of problems
            if "<" in problems_count:
                problems_count = 0
            else:
                problems_count = problems_count.replace(" ","") #format
                problems_count = [int(s) for s in problems_count.split() if s.isdigit()][0] #extract integer
        except IndexError:
            problems_count = 0
        
        publication_no = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/@id')[0] #define function that navigates to publication number
        date = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/ul/li/time/text()')[0] #navigates to date
        organisations1 = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/ul/li/abbr/@title') #navigates to first organisation
        organisations2 = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/ul/li[@class="organisations"]/text()') #navigates to second organisation
        organisations = organisations1 + organisations2

        organisations_abbrev1 = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']/ul/li/abbr') #navigates to organisation abreviation
        organisations_abbrev1 = [item.text for item in organisations_abbrev1] #list of abreviations
        organisations_abbrev2 = ['' * len(organisations2)]
        organisations_abbrev = organisations_abbrev1 + organisations_abbrev2

        pub_type = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']//li[@class="display-type"]/text()')[0] #navigates to publication type
        collections = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']//li[@class="document-collections"]/a') #navigates to collections
        collections = [item.text for item in collections] #list of collections
        collections_links = tree.xpath('//ol[@class="js-document-list document-list"]/li['+ str(j) +']//li[@class="document-collections"]/a/@href') #navigats to link of collection
        collections_links = ['https://www.gov.uk' + item for item in collections_links] #list of weblinks
        if len(collections) > 0:
            collection_views = [] #create list of collection views
            for webpage in collections_links:
                webpage = webpage.replace(".uk",".uk/info")
                test_page = requests.get(webpage)
                test_tree = html.fromstring(test_page.content) #define function that extracts the content from page
                
                view_count_col = test_tree.xpath('//p[@class = "count"]/text()')[0] #take first element of the list
                if "<" in  view_count_col:
                    view_count_col = 0
                else:
                    view_count_col = view_count_col.replace(" ","") #format
                    view_count_col = float(view_count_col.replace("\n","")) #format
                collection_views.append(view_count_col)

            collection_searches = [] #create list of collection searches
            for webpage in collections_links:
                webpage = webpage.replace(".uk",".uk/info")
                test_page = requests.get(webpage)
                test_tree = html.fromstring(test_page.content) #define function that extracts the content from page
                search_started_col = test_tree.xpath('//*[@id="lead-metrics"]/div[1]/div[2]/p/text()')[0] #take first element of the list
                search_started_col = search_started_col.replace(" ","") #format
                search_started_col = search_started_col.replace("\n","") #format
                collection_searches.append(search_started_col)

            collection_problems= []
            for webpage in collections_links:
                webpage = webpage.replace(".uk",".uk/info")
                test_page = requests.get(webpage)
                test_tree = html.fromstring(test_page.content) #define function that extracts the content from page
                problem_col = test_tree.xpath('//*[@id="lead-metrics"]/div[2]/div[1]/p/text()')[0] #take first element of the list
                problem_col = problem_col.replace(" ","") #format
                problem_col = problem_col.replace("\n","") #format
                collection_problems.append(problem_col)

            collection_stats = {"views": collection_views,
                               "searches": collection_searches, 
                               "problems": collection_problems} #create dictionary for collection statistics

            collection_stats['stats'] = [
                    {'views':i[0], 'searches':i[1],'problems':i[2]} 
                    for i in zip(collection_stats['views'], collection_stats['searches'], collection_stats['problems']) 
                ]
            del collection_stats['views']
            del collection_stats['searches']
            del collection_stats['problems']

            collection_statistics = collection_stats['stats']

            publication = {"Title": title,
                           "Publication_Stats":{"Unique_Views": view_count,
                                               "Searches_from_Page": search_count, 
                                               "Problems":problems_count}, 
                          "Link": link,
                          "Publication_Number": publication_no,
                          "Date": date,
                          "Organisation": organisations,
                          "Organisation_Abbreviations": organisations_abbrev,
                          "Publication_Type": pub_type,
                          "Collection": collections,
                          "Collection_Links": collections_links,
                          "Collection_Stats": collection_statistics} #create dictionary for publications

            publication['Organisations'] = [
                    {'Organisation':i[0], 'Organisation_Abbreviation':i[1]} 
                    for i in zip(publication['Organisation'], publication['Organisation_Abbreviations'])
                ]
            del publication['Organisation']
            del publication['Organisation_Abbreviations']

            publication['Collections'] = [
                    {'Collection':i[0], 'Collection_Link':i[1],"Collection_Stats":i[2]} 
                    for i in zip(publication['Collection'], publication['Collection_Links'],publication['Collection_Stats'])
                ]

            del publication['Collection']
            del publication['Collection_Links']
            del publication['Collection_Stats']

            publications.append(publication)
    
    print str(i) + " success"
    stop = timeit.default_timer()

    print (stop-start)/60
    print (stop-start)/page_num * noPages/60
    
publications
```

```python
#Save as JSON
with open('../data/govuk_publications4.json', 'w') as f:
     json.dump(publications, f)
```
